```python
# Dependencies and Setup
import pandas as pd
pd.options.display.float_format = '${:,.2f}'.format

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
# Displaying the total number of players
Player_Count= len(purchase_data["SN"].unique())
Player_Count_df=pd.DataFrame({"Player Count":[Player_Count]})
Player_Count_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Calculating Number of Unique Items
Number_Of_Unique_Items=len(purchase_data["Item ID"].unique())
# Calculating Average Price
Average_price=round(purchase_data["Price"].mean(),2)
# Calculating the Number of Purchases
Number_Of_Purchases=len(purchase_data["Purchase ID"].unique())
# Calculating the Total Revenue
Total_Revenue=purchase_data["Price"].sum()
# Putting everything into a dataframe
summary_df=pd.DataFrame({"Number Of Unique Items": [Number_Of_Unique_Items],
                          "Average price":Average_price,
                          "Number Of Purchases": Number_Of_Purchases,
                          "Total Revenue":Total_Revenue})
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed



```python
#Retrieving the SN and Gender into a Dataframe
Gender_df=purchase_data.loc[:, ["SN", "Gender"]]
# Removing the duplicates and assigning to a new dataframe
New_Gender_df=Gender_df.drop_duplicates()
# Calculating the Total_Count
Total_Count=len(New_Gender_df["Gender"])
# Determining the value counts
New_Gender_df["Gender"].value_counts()
# Assigning it to a new data frame with axis and index
Gender_Demo_df=New_Gender_df["Gender"].value_counts().rename_axis('Gender').reset_index(name='Total Count')
# Calculating the percentage of the Playerss
Gender_Percent = round(Gender_Demo_df["Total Count"]/Total_Count,4)*100
# Assigning the Percentage of the players to the Dataframe
Gender_Demo_df["Percentage of Players"] = Gender_Percent
# Formatting the Percentage of the players in the Dataframe
Gender_Demo_df["Percentage of Players"] = Gender_Demo_df['Percentage of Players'].astype(str) + '%'
Gender_Demo_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Grouping the purchase data by Gender and assigning to a new data frame
grouped_gender_df = purchase_data.groupby(["Gender"])
# Calculating the Purchase count from the grouped data frame and assigning to the count data frame
Purchase_Count_df=grouped_gender_df["Gender"].count().rename_axis('Gender').reset_index(name='Purchase Count')
Purchase_Count_df
# Calculating the Total Purchase Value from the grouped data frame and assigning to the purchase data frame
Total_Purchase_Value_df=grouped_gender_df["Price"].sum().rename_axis('Gender').reset_index(name='Total Purchase Value')
Total_Purchase_Value_df
# Merging the Purchase Count and the Total Purchase Value data frames
Merge_df=pd.merge(Purchase_Count_df,Total_Purchase_Value_df,on=["Gender"])
Merge_df
# Calaculating the Average Purchase price from the merged dataframe
Avg_Purchase_Price = round(Merge_df["Total Purchase Value"]/Merge_df["Purchase Count"],2)
# Assigning the Average Purchase Price to the merged data frame
Merge_df["Average Purchase Price"] = Avg_Purchase_Price
Merge_df

# From the previous cell Gender_Demo dataframe is considered for total count by Gender and sorting 
New_Gender_Demo_df=Gender_Demo_df.sort_values(["Gender"])
# Resetting the index
New_Gender_Demo_df=New_Gender_Demo_df.reset_index(drop=True)
New_Gender_Demo_df
# Calculating the Average Total Purchase per person
Avg_Tot_Pur_per_person= round(Merge_df["Total Purchase Value"]/New_Gender_Demo_df["Total Count"],2)
# Assigning the Average Total Purchase per person to the merged dataframe
Merge_df["Avg Total Purchase per Person"]=Avg_Tot_Pur_per_person
Merge_df
# Adding the Gender Index to the final dataframe
Final_Gender_df=Merge_df.set_index("Gender")
Final_Gender_df
# Re-arranging everything as per the Homework display
Final_Gender_df=Final_Gender_df[["Purchase Count","Average Purchase Price","Total Purchase Value","Avg Total Purchase per Person"]]
Final_Gender_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Retrieving the SN and Age values to Age_df dataframe
Age_df=purchase_data.loc[:, ["SN","Age"]]
Age_df
# Dropping the duplicates and assigning to a new dataframe
New_Age_df=Age_df.drop_duplicates()
New_Age_df
# Finding the total count of players in the dataframe
Total_Count=len(New_Age_df["SN"])
Total_Count
# Assigning the bins
bins=[0,9,14,19,24,29,34,39,44]
# Assigning the group names to be associated with the bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29","30-34","35-39",">40"]
# Assigning the distribution based off the bins and group names
New_Age_df["Age Distribution"] = pd.cut(New_Age_df["Age"], bins, labels=group_names, include_lowest=True)
New_Age_df
# Grouping the distribution by Age Distribution
Group_Age_Distribution_df = New_Age_df.groupby("Age Distribution")
Group_Age_Distribution_df.count()
# Calculatig the Total count for distributions
Age_Distribution_Summary_df=Group_Age_Distribution_df["Age Distribution"].count().rename_axis('Age Distribution').reset_index(name='Total Count')
Age_Distribution_Summary_df
# Calculating the percentage of the players based off the distribution and assigning to the dataframe
Percentage_of_Players=round(Age_Distribution_Summary_df["Total Count"]/Total_Count*100,2)
Age_Distribution_Summary_df["Percentage of Players"] = Percentage_of_Players
# Formatting the Distribution
Age_Distribution_Summary_df["Percentage of Players"] = Age_Distribution_Summary_df["Percentage of Players"].astype(str)+'%'
Age_Distribution_Summary_df
# Setting the index to the distribution
Final_Age_Distribution_Summary_df=Age_Distribution_Summary_df.set_index("Age Distribution")
Final_Age_Distribution_Summary_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Retrieving the needed columns SN and Price to a dataframe
Items_top_df=purchase_data.loc[:, ["SN","Price"]]
# Grouping the items by SN
Group_Items_top_df=Items_top_df.groupby(["SN"])
# Calculating the Total Purchase Value and assigning to a dataframe
Items_top_Price_df=Group_Items_top_df["Price"].sum().reset_index(name='Total Purchase Value')
Items_top_Price_df
# Calauclating the Purchase Count and assigning to a datarame
Items_top_count_df=Group_Items_top_df["SN"].size().reset_index(name='Purchase count')
Items_top_count_df
# Merging the Purchase COunt and the Total Purchase value data frames
merge_top_Items = pd.merge(Items_top_Price_df, Items_top_count_df, on=["SN"])
merge_top_Items
# Calculating the Average Purchase Price and adding to the dataframe
Average_Purchase_Price = round(merge_top_Items["Total Purchase Value"]/merge_top_Items["Purchase count"],2)
merge_top_Items["Average Purchase Price"]=Average_Purchase_Price
merge_top_Items
# Sorting the records on the Total Purchase Value to find the top spenders
Final_Sor_top_Merge=merge_top_Items.sort_values(["Total Purchase Value"],ascending=False)
# Setting the index to SN
Final_top_Merge=Final_Sor_top_Merge.set_index("SN")
Final_top_Merge.head(5)
# Arranging the records as requested in the homework
Final_top_Merge=Final_top_Merge[["Purchase count","Average Purchase Price","Total Purchase Value"]]
Final_top_Merge.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
